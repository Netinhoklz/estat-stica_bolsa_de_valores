# Importando dados

```python
import pandas as pd 
import MetaTrader5 as mt5
mt5.initialize()
data = pd.DataFrame(mt5.copy_rates_from_pos("WIN$N", mt5.TIMEFRAME_D1, 0, 252))
data
```

# Probabilidade de um dia ser ter a mesma direção (fechar positivo ou negativo) igual ao dia anterior

```python
data['alta/baixa'] = [1 if data['close'].iloc[x] - data['open'].iloc[x]>0 else 0 for x in range(data.shape[0])]
gain = 0
for k in range(1,data.shape[0]):
    if data['alta/baixa'][k] == data['alta/baixa'][k-1]:
        gain+=1
print(f'Acertos: {gain}\nErros: {data.shape[0]-gain}\nProbabilidade de acerto: {gain/data.shape[0]}')
```
