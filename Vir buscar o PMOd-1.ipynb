# Estudo sobre a probabilidade de ter um valor maior que a mínima do último

```python
import pandas as pd 
import MetaTrader5 as mt5
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
ativo = 'WIN$N'
mt5.initialize()
```

```python
#TESTE INDICADOR FAZER O LEVANTAMENTO ESTATÍSTICO DE QUANTAS VEZES ELE BATE A META MÍNIMA DOS ÚLTIMOS 21 DIAS DE VOLATILIDADE
ticks_dias = pd.DataFrame(mt5.copy_rates_from_pos('WIN$N', mt5.TIMEFRAME_W1, 0, 63))

ticks_dias['time'] = pd.to_datetime(ticks_dias['time'],unit = 's')

ticks_dias['pmo_d-1'] = (ticks_dias['high'].shift(1) + ticks_dias['low'].shift(1))/2

ticks_dias['buscou_pmo_d-1'] = [1 if ticks_dias['low'].iloc[x] < ticks_dias['pmo_d-1'].iloc[x] < ticks_dias['high'].iloc[x] else 0 for x in range(0,ticks_dias.shape[0])]

ticks_dias
```

```python
x = ticks_dias['buscou_pmo_d-1'].values
```

```python
#Probabilidade:
quantidade = np.sum(x == 1)
print(f'Probabilidade: {quantidade/len(x)}')
```
