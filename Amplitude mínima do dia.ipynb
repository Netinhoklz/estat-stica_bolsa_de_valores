# Estudo sobre a probabilidade de ter um valor maior que a mínima do último

```python
import pandas as pd 
import MetaTrader5 as mt5
from datetime import datetime
import matplotlib.pyplot as plt
ativo = 'WIN$N'
mt5.initialize()
```

```python
#TESTE INDICADOR FAZER O LEVANTAMENTO ESTATÍSTICO DE QUANTAS VEZES ELE BATE A META MÍNIMA DOS ÚLTIMOS 21 DIAS DE VOLATILIDADE
ticks_dias = pd.DataFrame(mt5.copy_rates_from_pos('WIN$N', mt5.TIMEFRAME_W1, 0, 52))
ticks_dias['time'] = pd.to_datetime(ticks_dias['time'],unit = 's')
ticks_dias['amplitude'] = ticks_dias['high'] - ticks_dias['low']
ticks_dias
```

```python
ticks_dias['min_21d'] = ticks_dias['amplitude'].rolling(window=5,closed='left').min()
ticks_dias.tail(50)
```

```python
x = tuple(ticks_dias[['amplitude','min_21d']].values)
contador_gain = 0
for i,y in x:
    if i > y:
        contador_gain += 1
        # print(i,y)
print(contador_gain)
```

```python
#Probabilidade:
contador_gain/len(ticks_dias['min_21d'])
```
