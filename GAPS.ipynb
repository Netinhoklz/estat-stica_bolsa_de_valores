<h1>Importações</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import MetaTrader5 as mt5
import datetime
import numpy as np
mt5.initialize()
ativo ='WIN$N'
periodo = 63
data = pd.DataFrame(mt5.copy_rates_from_pos(ativo, mt5.TIMEFRAME_D1, 1, periodo))
data
```

<h1>Probabilidade de fechar o GAP do dia anterior</h1>

```python
# for abertura,maxima,minima,fechamento in list(data[['open','high','low','close']].values):
#     if abertura > fechamento
total,_=data.loc[((data['open']>data['close'].shift(1))&(data['low']<data['close'].shift(1)))|((data['open']<data['close'].shift(1))&(data['high']>data['close'].shift(1)))].shape
total
```

```python
probabilidade = total/periodo
probabilidade
print(f'A probabildiade é de {round(probabilidade*100,2)}%, com base nos últimos {periodo} pregões.')
```
