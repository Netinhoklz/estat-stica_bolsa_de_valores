<h1>Importações</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import MetaTrader5 as mt5
import datetime
import numpy as np
mt5.initialize()
ativo ='WDO$N'
data_10m = pd.read_excel('enor_grafico_10m.xlsx')
data = pd.DataFrame(pd.read_excel('graf_diario.xlsx'))
data['Amplitude_Vwap'] = abs(data['VWAP D'] - data['VWAP D'].shift(-1))
data['Amplitude'] = data['Máxima'] - data['Mínima']
data['media_amplitude'] = data['Amplitude'].rolling(22).mean()
data_10m
```

<h1>Probabilidade dele romper o candle de 10m para cima e para baixo </h1>

```python
#Probabilidade dele romper o candle de 10m para cima e para baixo
contador_diario = 0
contador_10m = 0
rompeu_cima_e_baixo = 0
for data_9 in data_10m['Data']:
    for data_d in data['Data'][0:200]:
        if np.datetime64(data_9,'D') == np.datetime64(data_d,'D'):
            if data['Máxima'][contador_diario] > data_10m['Máxima'][contador_10m] and data['Mínima'][contador_diario] < data_10m['Mínima'][contador_10m]:
                rompeu_cima_e_baixo += 1
        contador_diario +=1
    contador_diario = 0 
    contador_10m+=1
print(rompeu_cima_e_baixo,rompeu_cima_e_baixo/181)
```
